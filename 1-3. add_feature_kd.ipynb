```python
def add_feature(pre, feature):
    read_feature = pd.read_csv('C:/Users/dladu/Desktop/위암 project/5차modeling/기저질환/'+feature+'.csv')
    feature_df = pd.DataFrame(read_feature)
    feature_df.columns = ['person_id','person_id2','concept_id','date']
    del feature_df['person_id2']
    
    feature_df.sort_values(by=['person_id','date'],axis=0,inplace=True)
    feature_df = feature_df.drop_duplicates(['person_id'],keep='first')
    feature_df.reset_index(drop=True, inplace=True)
    #print(feature_df)
    
    p_ids = list(feature_df['person_id'])
    #print(p_ids)
    for i in range(len(pre)):
        #print(pre.iloc[i,0])
        if pre.iloc[i,0] in p_ids:
            p_index = p_ids.index(pre.iloc[i,0])
            feature_date = (pd.to_datetime(feature_df.iloc[p_index,2])).year
            pre_date = (pd.to_datetime(pre.iloc[i,7])).year
            if pre_date<feature_date : # 최초 진료 기록이 그 이후일 경우
                pre.loc[i,feature] = 0
            else :
                pre.loc[i,feature] = 1
        if i%5000==0:
            #pass
            print(i)
    if feature in pre.columns:
        print(feature+' 입력값 개수 : {:d} /323'.format(pre[feature].count()))
    
    return pre
```
