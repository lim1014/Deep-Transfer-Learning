```python
import pandas as pd
import numpy as np
from collections import Counter
from sklearn.metrics import confusion_matrix,classification_report
```

```python
pre = pd.read_csv('final_0901.csv')

pre_df = pd.DataFrame(pre)

pre_df
```

```python
del pre_df['date']
del pre_df['observation_date']
```

```python
# age, gender 포함 기저질환만 사용

pre_df = pre_df[['person_id','gender', 'age', 'bmi','smoking_status',
       'Malignant neoplastic disease', 'Chronic liver disease',
       'chronic obstructive lung disease', 'cerebrovascular disease',
       'chronic kidney disease', 'Diabetes mellitus', 'Ischemic heart disease',
       'hyperlipidemia', 'Hypertensive disorder', 'cancer']]
```

```python
pre_df.loc[:,list(pre_df.loc[:,pre_df.isnull().any()].columns)].isnull().sum()/(len(pre_df))*100
```

```python
# 범주형 변수의 Unique 값 살펴보기
cat_col =  ['Malignant neoplastic disease', 'Chronic liver disease',
       'chronic obstructive lung disease', 'cerebrovascular disease',
       'chronic kidney disease', 'Diabetes mellitus', 'Ischemic heart disease',
       'hyperlipidemia', 'Hypertensive disorder']
for col in cat_col:
    print(pre_df.loc[:,col].unique())
```

```python
pre_df.fillna({'bmi':round(pre_df['bmi'].mean(),2),
                'smoking_status':0,
                'Malignant neoplastic disease':0,
                'Chronic liver disease':0,
                'chronic obstructive lung disease':0,
                'cerebrovascular disease':0,
                'chronic kidney disease':0,
                'Diabetes mellitus':0,
                'Ischemic heart disease':0,
                'hyperlipidemia':0,
                'Hypertensive disorder':0},inplace=True)
```

```python
pre_df.loc[:,list(pre_df.loc[:,pre_df.isnull().any()].columns)].isnull().sum()/(len(pre_df))*100
# 결측 확인
```

```python
#pre_df.to_csv('graph_data_kbd.csv', mode='w', index=False)
```

```python
pre_df['bmi'].describe()
```

```python
# bmi 18.5이하 저체중, 23이하 정상, 25이하 과체중

pre_df.loc[pre_df['bmi']<=18.5, 'bmi'] = 1
pre_df.loc[((pre_df['bmi']<=23)&(pre_df['bmi']>18.5)), 'bmi'] = 2
pre_df.loc[((pre_df['bmi']<=25)&(pre_df['bmi']>23)), 'bmi'] = 3
pre_df.loc[(pre_df['bmi']>25), 'bmi'] = 4
```

```python
# 나이 10살 단위 범주화

# 어린 사람 ~ 나이 많은 사람 순

pre_df.loc[pre_df['age']<=20, 'age'] = 1
pre_df.loc[((pre_df['age']<=30)&(pre_df['age']>20)), 'age'] = 2
pre_df.loc[((pre_df['age']<=40)&(pre_df['age']>30)), 'age'] = 3
pre_df.loc[((pre_df['age']<=50)&(pre_df['age']>40)), 'age'] = 4
pre_df.loc[((pre_df['age']<=60)&(pre_df['age']>50)), 'age'] = 5
pre_df.loc[((pre_df['age']<=70)&(pre_df['age']>60)), 'age'] = 6
pre_df.loc[((pre_df['age']<=80)&(pre_df['age']>70)), 'age'] = 7
pre_df.loc[(pre_df['age']>80), 'age'] = 8
```

```python
print(pre_df.loc[:,'age'].unique())
```

```python
# 범주형이 된 age의 가변수 생성
new_pre = pd.get_dummies(pre_df, columns=['age','smoking_status','bmi'])
```

```python
new_pre.shape
```

```python
new_pre = new_pre[['person_id','gender', 'age_1', 'age_2', 'age_3', 'age_4',
       'age_5', 'age_6', 'age_7', 'age_8','smoking_status_0.0',
       'smoking_status_1.0', 'smoking_status_2.0', 'smoking_status_3.0',
       'bmi_1.0', 'bmi_2.0', 'bmi_3.0', 'bmi_4.0',
       'Malignant neoplastic disease', 'Chronic liver disease',
       'chronic obstructive lung disease', 'cerebrovascular disease',
       'chronic kidney disease', 'Diabetes mellitus', 'Ischemic heart disease',
       'hyperlipidemia', 'Hypertensive disorder', 'cancer']]
```
