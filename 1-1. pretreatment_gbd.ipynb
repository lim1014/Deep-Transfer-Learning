## 음주습관

```python
al = pd.read_csv('alcohol_habit0728.csv')

al_df = pd.DataFrame(al)

al_df.sort_values(by=['person_id','observation_date'],axis=0,inplace=True)
```

```python
# 1)	(거의)마시지 않는다     2) 월2~3회정도 마신다 3) 일주일에 1~2회 마신다  
# 4) 일주일에 3~4회 마신다  5) 거의 매일 마신다

al_df.loc[(al_df['value_as_concept_id']==45882527),['observation_source_value']]=1
al_df.loc[(al_df['value_as_concept_id']==45885249),['observation_source_value']]=1
al_df.loc[(al_df['value_as_concept_id']==45881653),['observation_source_value']]=2
al_df.loc[(al_df['value_as_concept_id']==45885248),['observation_source_value']]=3
al_df.loc[(al_df['value_as_concept_id']==45879676),['observation_source_value']]=4
```

```python
# 1)	0회 음주 2번) 1회 음주 3번) 2회 음주 ……8번) 7회 음주 
al_df.loc[(al_df['unit_concept_id']==45881908) & (al_df['observation_source_value']==3),['observation_source_value']]=2
al_df.loc[(al_df['unit_concept_id']==45881908) & (al_df['observation_source_value']==4),['observation_source_value']]=3
al_df.loc[(al_df['unit_concept_id']==45881908) & (al_df['observation_source_value']==5),['observation_source_value']]=3
al_df.loc[(al_df['unit_concept_id']==45881908) & (al_df['observation_source_value']==6),['observation_source_value']]=4
al_df.loc[(al_df['unit_concept_id']==45881908) & (al_df['observation_source_value']==7),['observation_source_value']]=4
al_df.loc[(al_df['unit_concept_id']==45881908) & (al_df['observation_source_value']==8),['observation_source_value']]=4
```

```python
Counter(al_df['observation_source_value'])
```

```python
al_df
```

```python
al_df = al_df.loc[:,['person_id','observation_date','observation_source_value']]
al_df.columns = ['person_id','observation_date','alcohol_habit']
```

```python
treat = pd.merge(real_cg,al_df,how='left',on=['person_id'])
```

```python
treat
```

```python
#Counter(treat['alcohol_habit'])
```

```python
treat.loc[:,list(treat.loc[:,treat.isnull().any()].columns)].isnull().sum()/(len(treat))*100
```

## BMI

```python
bmi = pd.read_csv('measurement0728.csv')

bmi_df = pd.DataFrame(bmi)
```

```python
waist = bmi_df[bmi_df['measurement_concept_id']== 3016258].index
bmi_df = bmi_df.drop(waist)
```

```python
bmi_df.sort_values(by=['person_id','measurement_date'])
```

```python
bmi_df.loc[:,list(bmi_df.loc[:,bmi_df.isnull().any()].columns)].isnull().sum()/(len(bmi_df))*100
# 결측 확인
```

```python
height = bmi_df[bmi_df['measurement_concept_id']==3036277]
weight = bmi_df[bmi_df['measurement_concept_id']==3025315]
```

```python
del height['measurement_concept_id']

height.columns = ['person_id','observation_date','height']

del weight['measurement_concept_id']

weight.columns = ['person_id','observation_date','weight']

bmi_merge = pd.merge(height,weight,how = 'outer',on=['person_id','observation_date'])
bmi_merge['bmi'] = round(bmi_merge['weight']/((bmi_merge['height']/100)**2),2)
bmi_merge = bmi_merge.dropna(axis=0)
```
